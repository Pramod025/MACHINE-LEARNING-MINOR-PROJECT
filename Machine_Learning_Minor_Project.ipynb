```python
x = Data.iloc[:,0 : 3].values
y = Data.iloc[:, -1].values

x
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.8,random_state=0)
x_train
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
print('"successfully trained the Model"')
```

```python
y_predict = model.predict(x_test)
y_predict
```

```python
testing_data_model_score = model.score(x_test, y_test)
print("Model Score on Testing data :",testing_data_model_score)

training_data_model_score = model.score(x_train, y_train)
print("Model Score on Training data :",training_data_model_score)
```

```python
df = pd.DataFrame(data={'Predicted value':y_predict.flatten(),'Actual Value':y_test.flatten()})
df
```

```python
df.plot(figsize=(15,8))
plt.show()
```

```python
df.plot(figsize=(15,8),kind = 'bar')
```

```python
sns.regplot(x='Actual Value',y='Predicted value',data=df)
```

```python
from sklearn.metrics import r2_score

r2Score = r2_score(y_test, y_predict)
print("R2 score of model is :" ,r2Score)
```
